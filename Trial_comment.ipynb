```python
import facebook_scraper as fs
import pandas as pd
# get POST_ID from the URL of the post which can have the following structure:
# https://www.facebook.com/USER/posts/POST_ID
# https://www.facebook.com/groups/GROUP_ID/posts/POST_ID
#POST_ID = "pfbid02gqM2rFz62gqaGPxhQQJb64koXQskRstiNYLKFDRM197L5rSSQKK8r3QU6n4MZMnKl"
POST_ID = "1641501043032205"
# number of comments to download -- set this to True to download all comments
MAX_COMMENTS = 100
# get the post (this gives a generator)
gen = fs.get_posts(
    post_urls=[POST_ID],
    options={"comments": MAX_COMMENTS, "progress": True}
)
# take 1st element of the generator which is the post we requested
post = next(gen)
# extract the comments part
comments = post['comments_full']
commant = []
for comment in comments:
    commant.append(comment)

    
    
    
dat = pd.DataFrame(commant)    
dat
    
```

```python
d= dat['replies']
new = dat[['comment_id','commenter_id','comment_text']]
replies = []
for k in range(len(d)):
    replies.append([i['comment_text'] for i in d[k]])
new['replies']=replies  
```

```python
new['replies']= new['replies'].apply(lambda x:"N" if not x else x)
new['count']=new['replies'].apply(lambda x: len(x))
```

```python
commentor = []
for k in range(len(d)):
    commentor.append([i['commenter_name'] for i in d[k]])
new['commentor']=commentor 
new['commentor']= new['commentor'].apply(lambda x:"N" if not x else x)
new['cont']=new['commentor'].apply(lambda x: len(x))
```

```python
now_rows = []
for i in range(new.shape[0]):
    now_rows.append(pd.concat([new.loc[i]]*new['count'].loc[i], axis = 1).transpose())
```

```python
Comments = pd.concat(now_rows)
Comments
Reply = [r for y in new['replies'] for r in y]
comentor = [r for y in new['commentor'] for r in y]
Comments['Reply']= Reply
Comments['Commentor']= comentor
Comments = Comments[['comment_id','commenter_id','comment_text','Reply','Commentor']]
Comments
```
